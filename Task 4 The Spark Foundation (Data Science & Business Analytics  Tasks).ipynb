```python
# let's represent the correlation in the form of graph

global_t.hist(figsize = (50,55))

plt.show()
```

```python
# let's print region terrorist activity in each yeaer using count plot

sns.set(rc={'figure.figsize': (25,40)})
sns.countplot(x=global_t["iyear"], hue=global_t["region_txt"])
plt.title("Region-wise Terrorist Activity in Each Year")
```

```python
# to understand and explore in deep lets try to visualize it with cross tab

pd.crosstab(global_t['iyear'], global_t['region_txt']).plot(kind="line",figsize=(15,20))
plt.title("Region-wise Terrorist Activity in Each Year")
plt.ylabel("Number of Attacks")
plt.xlabel("Years")
plt.show()
```

### From 2001 global terrorism has increased so badly. major impact took place in Middle east and North Africa region.

```python
# let's find out most affected region

global_t["region_txt"].value_counts()  # Ass per the data Middle East & North Africa are most affected 
```

```python
# let's print most affected countires 

global_t["country_txt"].value_counts()     # country wise Iraq, pakistan and Afghanistan are most affected countires.
```

```python
# let's print top 15 countires affected by Terrorism

plt.figure(figsize=(15,6))
sns.barplot(global_t["country_txt"].value_counts()[:15].index, global_t["country_txt"].value_counts()[:15].values)
plt.title("Top 15 Countries Affected by Attacks", fontsize=15)
plt.ylabel("Counts", fontsize = 15)
plt.xlabel("Countries", fontsize = 15)
plt.show()


print("Most Affected Countries are Ireq and Pakistan.")
```

```python
# Let's also find out most affected states.

plt.figure(figsize = (15,6))
sns.barplot(global_t["city"].value_counts()[:5].index, global_t["city"].value_counts()[:5].values)
plt.title("Top 5 Cities Affected by Attacks", fontsize = 15)
plt.ylabel("Counts", fontsize = 15)
plt.xlabel("City", fontsize = 15)

plt.show()
```

### Most Affected City Name is not mentioned and the second most affected is BAGHDAD.

```python
# let's find out weapon type used for terrorism attacks

sns.set(rc = {"figure.figsize": (30,7)})
sns.countplot(x= global_t["region_txt"], hue = global_t["weaptype1_txt"])
plt.title("Weapon Type used in Per Region")
plt.xlabel("Region", fontsize = 15)
```

### Explosives, Chemicals and Firearms are widely used as a weapon type to destroy harmony.

```python
# let's visualize what are the attack types has been used in Middle East and North Africa

sns.set(rc = {"figure.figsize": (30,7)})
sns.countplot(x= global_t["region_txt"], hue = global_t["attacktype1_txt"])
plt.title("Attack Type used in Per Region")
plt.xlabel("Region", fontsize = 15)
```

### Bombing and Explosion is highly used in Middle Ease and North Africa Region.

```python
# let's find out property types where attacks took place in huge numbers

sns.set(rc = {"figure.figsize": (30,7)})
sns.countplot(x= global_t["region_txt"], hue = global_t["targtype1_txt"])
plt.title("Most Affected Property Type")
plt.xlabel("Region", fontsize = 15)
```

### Private property has been targated mostly by the terrorist. and colateral damage as per the graph is devastating.

```python
# let's understand total number of death and injuries in overall.

global_t.plot(kind="scatter", x = "nkill", y = "nwound", color = "red", figsize = (10,10),fontsize = 15)
plt.xlabel("Kill")
plt.ylabel("Wound")
plt.title("Killed - Wounded Scatter Plot")
plt.show()
```
